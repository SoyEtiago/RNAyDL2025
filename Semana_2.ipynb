# **Definición de red neuronal**

Es un modelo inspirado en el funcionamiento del cerebro humano, diseñado para simular la forma en que los humanos aprenden. Está compuesta por unidades de procesamiento llamadas neuronas, organizadas en capas, a continuación se muestra un modelo biologico de neurona y un modelo matematico de neurona (tambien conocido como [*Perceptron*](https://towardsdatascience.com/an-introduction-to-perceptron-algorithm-40f2ab4e2099))

# **Componentes de una neurona artificial**

La unidad mas pequeña de una red neuronal es la neruona, esta se encuentra compuesta por disintas partes las cuales se menciona a continuación:

**Entradas**

Son los datos que se proporcionan al modelo para su procesamiento y análisis. Son el punto de partida para lo que la red neuronal intentará aprender, clasificar, predecir o reconocer a través de su estructura de capas y neuronas. Estas entradas pueden venir en diversas formas, dependiendo del tipo de problema y los datos disponibles.

1.   **Datos Numéricos:** Las entradas pueden ser vectores de números que representan características cuantitativas de los ejemplos de datos. Por ejemplo, en una tarea de predicción del precio de casas, las entradas podrían incluir el tamaño de la casa, el número de habitaciones, la edad de la propiedad, etc.

2.   **Imágenes:** En tareas de visión por computadora, como el reconocimiento de imágenes, las entradas son típicamente las intensidades de píxeles de imágenes. Estas imágenes son transformadas en matrices de números, donde cada número representa la intensidad de un píxel en la imagen.

3.   **Texto:** Para el procesamiento del lenguaje natural (PLN), las entradas son textos que deben ser convertidos a formatos numéricos que la red pueda procesar. Esto se hace a menudo a través de técnicas como la codificación one-hot, embeddings de palabras, o transformaciones más complejas que capturan el significado semántico de las palabras y oraciones.

4.   **Señales de Audio:** En aplicaciones de reconocimiento de voz o música, las entradas son señales de audio digitalizadas. Estas señales pueden ser transformadas en espectrogramas o características de Mel-frequency cepstral coefficients (MFCC) que representan aspectos relevantes del sonido para el aprendizaje.

5.   **Datos de Series Temporales:** Para predicciones basadas en el tiempo, como el pronóstico del mercado de valores o la meteorología, las entradas son series de datos temporales. Estos pueden incluir secuencias de precios históricos, mediciones meteorológicas a lo largo del tiempo, etc.

**Pesos**

Son parámetros fundamentales que determinan la fuerza y el signo de la conexión entre neuronas y la entrada. En esencia, los pesos controlan la influencia que tiene la entrada en una neurona (o varias neuronas si hablamos de que antes habia una capa) sobre la señal de salida que produce.

Cada conexión entre neuronas en una red neuronal tiene asociado un peso único. Cuando una señal pasa de una neurona a otra, se multiplica por el peso de la conexión. Este proceso es una parte crucial de cómo las redes neuronales procesan la información y aprenden patrones complejos de los datos.

image.png


Al igual que las entradas tabulares, los pesos se suelen representar en matrices


**Sumatoria**

En el contexto de un perceptrón, la sumatoria de las entradas con los pesos es un paso fundamental en su operación. Esta sumatoria, seguida de la aplicación de una función de activación, determina si la neurona se activa o no, es decir, si envía una señal hacia adelante o no.

El perceptrón recibe múltiples entradas (denotadas como \(x_1, x_2, ..., x_n\)), y cada entrada está asociada con un peso específico (denotados como \(w_1, w_2, ..., w_n\)). La suma ponderada de estas entradas es la base para el cálculo que realiza el perceptrón. Además, a menudo se incluye un término de sesgo (\(b\)), que permite ajustar el umbral de activación de la neurona.

La fórmula matemática para la sumatoria de las entradas ponderadas por sus pesos, incluyendo el sesgo, es la siguiente:

\[Neta = w_1x_1 + w_2x_2 + ... + w_nx_n + b\]

Donde:
- \(Neta\) es el resultado de la suma ponderada y el sesgo.
- \(w_i\) representa el peso asociado a la entrada \(i\).
- \(x_i\) es el valor de la entrada \(i\).
- \(b\) es el sesgo.

image.png

```python
# Matriz de entrada X con dimensiones 3x5 (3 filas, 5 columnas)
# Cada fila representa una observación diferente
# Cada columna representa una característica diferente
X = np.array([[1, 2, 9, 10,1],
              [3, 4, 1, 1,1],
              [5, 6, 3, 7,1]])

# Matriz de pesos W con dimensiones 5x2 (5 filas, 2 columnas)
# Esto significa que transformaremos nuestras 5 características de entrada
# en 2 nuevas características (o salidas) mediante la multiplicación de matrices

W = np.array([[0.5, -0.5],
              [-0.5, 0.5],
              [0.8, 0.8],
              [0.8, 0.8],
              [0.8, 0.8]])

# Multiplicación de la matriz de entrada por la matriz de pesos
# Resultado tendrá dimensiones 3x2
Neta = np.dot(X, W)

print("Matriz de entrada X:")
print(X)
print("\nMatriz de pesos W:")
print(W)
print("\nResultado de la multiplicación Neta = X*W:")
print(Neta)
```

**Función de Activación**

Después de calcular la suma ponderada \(Neta\), el perceptrón aplica una función de activación para determinar la salida. En el modelo más simple de perceptrón, esta función de activación suele ser una función escalón, también conocida como función de umbral, que se define como:

\[f(Neta) = \begin{cases}
1 & \text{si } Neta \geq \text{umbral} \\
0 & \text{en otro caso}
\end{cases}\]

En este caso, el "umbral" se ajusta implícitamente con el valor del sesgo \(b\). Si la suma ponderada \(Neta\) alcanza o supera este umbral, el perceptrón se activa (envía una señal de salida de 1); de lo contrario, no se activa (la señal de salida es 0).

A continuación se muestran otro tipo de funcionés de activación, **recuerda que para dicha tabla X representa la sumatorio (Neta) y  f(x) representa a la salida de la neurona**:

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """Función de activación Sigmoid."""
    return 1 / (1 + np.exp(-x))

def tanh(x):
    """Función de activación Tanh."""
    return np.tanh(x)

def relu(x):
    """Función de activación ReLU."""
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    """Función de activación Leaky ReLU."""
    return np.where(x > 0, x, x * alpha)

def softmax(x):
    """Función de activación Softmax."""
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)


# Generar un rango de valores para demostrar las funciones de activación
x = np.linspace(-10, 10, 100)

# Aplicar las funciones de activación
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_softmax = softmax(np.array([x]))  # Softmax se aplica típicamente a lo largo de un eje para vectores

# Visualizar las funciones de activación
plt.figure(figsize=(14, 9))

# Sigmoid
plt.subplot(2, 3, 1)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.title("Sigmoid")
plt.xlabel("Neta")
plt.ylabel("Salida")
plt.grid(True)

# Tanh
plt.subplot(2, 3, 2)
plt.plot(x, y_tanh, label="Tanh")
plt.title("Tanh")
plt.xlabel("Neta")
plt.ylabel("Salida")
plt.grid(True)

# ReLU
plt.subplot(2, 3, 3)
plt.plot(x, y_relu, label="ReLU")
plt.title("ReLU")
plt.xlabel("Neta")
plt.ylabel("Salida")
plt.grid(True)

# Leaky ReLU
plt.subplot(2, 3, 4)
plt.plot(x, y_leaky_relu, label="Leaky ReLU")
plt.title("Leaky ReLU")
plt.xlabel("Neta")
plt.ylabel("Salida")
plt.grid(True)

# Softmax
plt.subplot(2, 3, 5)
for i in range(y_softmax.shape[0]):
    plt.plot(x, y_softmax[i], label=f"Softmax {i+1}")
plt.title("Softmax")
plt.xlabel("Neta")
plt.ylabel("Salida")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Realice la función de activación en codigo teniendo como referencia un umbral de 0, es decir si la neta tiene un valor mayor a 0 su salida sera 1 y si la neta tiene un valor menor a 0 la salida sera -1

A continuación se muestra que la red  neuronal se compone de una capa de entrada, una o varias capas ocultas y la capa de salida. Tanto la capa oculta como la capa de salida tiene operaciones de entrada, la capa de entrada NO

```python
def inicializacion_capa_oculta(Numero_entradas, bias):

    Neurona_1 = np.random.randn(Numero_entradas + bias, 1)

    Neurona_2 = np.random.randn(Numero_entradas + bias, 1)

    Neurona_3 = np.random.randn(Numero_entradas + bias, 1)

    Neurona_4 = np.random.randn(Numero_entradas + bias, 1)

    parameters = {"Pesos_Neurona_1": Neurona_1, "Pesos_Neurona_2": Neurona_2, "Pesos_Neurona_3": Neurona_3, "Pesos_Neurona_4": Neurona_4}


    return parameters

parameters = inicializacion_capa_oculta(2,1)
```

**Feedforward**


Es el proceso de propagar las entradas a través de la red, capa por capa, desde la capa de entrada hasta la capa de salida, sin ciclos ni conexiones hacia atrás. Este proceso utiliza las operaciones de suma ponderada y las funciones de activación.

Durante el feedforward, cada neurona en una capa recibe entradas de las neuronas de la capa anterior (o de los datos de entrada en el caso de la capa de entrada), calcula una suma ponderada de estas entradas (incluyendo un término de sesgo), aplica una función de activación a este sumatorio, y luego envía el resultado a las neuronas de la siguiente capa. Este proceso se repite hasta que las señales han pasado por todas las capas de la red y llegan a la capa de salida, donde se produce el resultado final del modelo.

```python
# Propagación hacia adelante para la capa ocultada

def forward_pass_capa_oculta(X, parameters):
    # PN1 = PESOS NEURONA 1

    PN1, PN2, PN3, PN4 = parameters.values()

    # Calculo de las Netas y salidas

    Neta_1 = np.dot(X,PN1)
    Salida_1 = sigmoid(Neta_1)

    Neta_2 = np.dot(X,PN2)
    Salida_2 = sigmoid(Neta_2)

    Neta_3 = np.dot(X,PN3)
    Salida_3 = sigmoid(Neta_3)

    Neta_4 = np.dot(X,PN4)
    Salida_4 = sigmoid(Neta_4)

    Salidas = {"Salida_1": Salida_1, "Salida_2": Salida_2, "Salida_3": Salida_3, "Salida_4": Salida_4}

    return Salidas

Salidas_capa_oculta = forward_pass_capa_oculta(X, parameters)
```

```python
def inicializacion_capa_salida(Numero_entradas, bias):

    Neurona_salida = np.random.randn(Numero_entradas + bias, 1)

    parameter_2 = {"Pesos_neurona_salida": Neurona_salida}

    return parameter_2

parameter_2 = inicializacion_capa_salida(4, 1)
```

```python
# Propagación hacia adelante para la capa ocultada

def forward_pass_capa_oculta(Salidas_capa_oculta, parameter_2):
    # S1 = SALIDA NEURONA 1

    S1, S2, S3, S4 = Salidas_capa_oculta.values()

    # Calculo de las Netas y salidas

    Almacenamiento_salidas = []

    for i in range(4):

      Entradas_capa_salida = np.array([[S1[i][0]],[S2[i][0]],[S3[i][0]],[S4[i][0]],[1]])

      Netas = np.dot(Entradas_capa_salida.T,list(parameter_2.values())[0])

      Salida = sigmoid(Netas)

      Almacenamiento_salidas.append(Salida)

    return Almacenamiento_salidas

Salidas_capa_salida = forward_pass_capa_oculta(Salidas_capa_oculta, parameter_2)
```
