```python
# @title ENTEROS
#Puedo asignar el valor a una variable por medio del =
nombre = 10
entero_1 = 418

entero_2 =10

suma = entero_1 + entero_2

print("La suma del entero 1 y entero 2 es: ", suma)

resta = entero_1 - entero_2

print("La resta del entero 1 y entero 2 es: ", resta)
```

```python
# @title VARIABLES BOOLEANAS
bol_1 = True

bol_2 = True

print("El resultado de la variable 1 Y la variable 2 es: ", bol_1 and bol_2)

print("El resultado de la variable 1 O la variable 2 es: ", bol_1 or bol_2)

print("El resultado negado de la variable 1 es: ", not bol_1)
```

# FUNCIONES

```python
def HelloWorldXY(x, y):
  if (x < 10):
    print("Hello World, x was < 10")
  elif (x < 20):
    print("Hello World, x was >= 10 but < 20")
  else:
    print("Hello World, x was >= 20")
  return x + y

for i in range(8, 25, 5):  # i=8, 13, 18, 23 (start, stop, step)
  print("--- Now running with i: {}".format(i))
  r = HelloWorldXY(i,i)
  print("Result from HelloWorld: {}".format(r))
```
